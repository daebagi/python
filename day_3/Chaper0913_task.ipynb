__[문제3]<br>__

다음 리스트에서 최댓값과 최솟값, 그리고 평균을 출력해주세요.
<br>

```
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]
```



````python
# min max ? 강의에 없던데용 파이썬에는 average함수가 없다...

nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]

최댓값 = max(nums)
최솟값 = min(nums)
평균 = sum(nums) / len(nums)

print(f'최댓값 : {최댓값}')
print(f'최솟값 : {최솟값}')
print(f' 평균 : {평균}')
````
